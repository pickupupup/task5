## 重排与分级排序

swaplevel() 接受两个级别编号或名称，并返回一个互换了级别的新的对象

```python
frame
```

```python
frame.swaplevel('key1','key2')
```

sort_index()根据单个级别中的值对数据进行排序。交换级别时，常常也会用到sort_index(),这样最终结构就是按照指定顺序进行字母排序

```python
frame.sort_index(level=1)
```

```python
frame.swaplevel(0,1).sort_index(level=0)
```

## 数据库风格的DF合并

```python
df1=pd.DataFrame({'key':['b','b','a',
                        'c','a','a','b'],
                'data1':range(7)})
df2=pd.DataFrame({'key':['a','a','d'],
                 'data2':range(3)})
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1,df2,on='key')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
df3=pd.DataFrame({'key3':['b','b','a',
                        'c','a','a','b'],
                'data1':range(7)})
df4=pd.DataFrame({'key4':['a','a','d'],
                 'data2':range(3)})
pd.merge(df3,df4,left_on='key3',right_on='key4')
```

```python
df1=pd.DataFrame({'key':['b','b','a',
                        'c','a','b'],
                'data1':range(6)})
df2=pd.DataFrame({'key':['a','b','a','b','d'],
                 'data2':range(5)})
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='key',how="left")
```

```python
pd.merge(df1,df2,how="inner")
```

```python
left=pd.DataFrame({'key1':['foo','foo','bar'],
                     'key2':['one','two','one'],
                     'lval':[1,2,3] })
right=pd.DataFrame({'key1':['foo','foo','bar','bar'],
                   'key2':['one','one','one','two'],
                   'rval':[4,5,6,7]})
pd.merge(left, right, on=['key1','key2'],how='outer')
# out 使用表中所有的键
```

```python
pd.merge(left,right,on='key1')
```

```python
pd.merge(left,right,on='key1',suffixes=('_left','_right'))
```

## 重塑层次化索引

stack 列转为行

unstack 行转为列

```python
data=pd.DataFrame(np.arange(6).reshape((2,3)),
                 index=pd.Index(['ohio','colorado'],
                               name='state'),
                               columns=pd.Index(['one','two',
                                                 'three'],
                                                name='number'))
data
```

```python
result=data.stack()
result
```

```python
result.unstack()
```

```python
result.unstack(0)
```

```python
result.unstack('state')
```

```python
s1=pd.Series([0,1,2,3],index=['a','b','c','d'])
s2=pd.Series([4,5,6],index=['c','d','e'])
data2=pd.concat([s1,s2],keys=['one','two'])
data2
```

```python
data2.unstack()
```

```python
data2.unstack().stack()
```

```python
data2.unstack().stack(dropna=False)
```

```python
df=pd.DataFrame({'left':result,
                'right':result+5},
               columns=pd.Index(['left','right'],name='side'))
df
```

```python
df.unstack('state')
```

```python
df.unstack('state').stack('side')
```
